```python
df["Population"] = df.Population.astype(float)
df["latitude"] = df.latitude.astype(float)
df["longitude"] = df.longitude.astype(float)
```

```python
#Create the map for first dose
v_map = folium.Map(location=[21, 78], tiles='cartodbpositron', zoom_start=4)
#Add points to the map
mc = MarkerCluster()
for idx, row in df.iterrows(): 
    if not math.isnan(row['longitude']) and not math.isnan(row['latitude']):
        mc.add_child(Marker(location=[row['latitude'], row['longitude']],
                            tooltip=str(round(row['percentage']*100, 2))+"%"))
        
v_map.add_child(mc)

# Display the map
v_map
```

```python
#Create the map for second dose
v_map = folium.Map(location=[21, 78], tiles='cartodbpositron', zoom_start=4)
#Add points to the map
mc = MarkerCluster()
for idx, row in df.iterrows(): 
    if not math.isnan(row['longitude']) and not math.isnan(row['latitude']):
        mc.add_child(Marker(location=[row['latitude'], row['longitude']],
                            tooltip=str(round(row['percentage2']*100, 2))+"%"))
        
v_map.add_child(mc)

# Display the map
v_map
```
